## 2. Vectorised Operations — No Loops

The core principle of NumPy: **never write a for-loop over array elements**. NumPy operations run in optimised C, so vectorised code is 10-100x faster than Python loops.

Element-wise operations apply to every element simultaneously:

````python
# Example: element-wise operations
x = np.array([1.0, 2.0, 3.0, 4.0])

print(f"x + 10   = {x + 10}")      # add scalar to every element
print(f"x * 2    = {x * 2}")       # multiply every element
print(f"x ** 2   = {x ** 2}")      # square every element
print(f"np.exp(x)= {np.exp(x)}")   # e^x for every element
print(f"np.log(x)= {np.log(x)}")   # ln(x) for every element
````

## 3. Broadcasting — The Tricky Part

Broadcasting is how NumPy handles operations between arrays of **different shapes**. It's the most common source of bugs AND the most powerful feature for writing clean ML code.

**The rules:**
1. If arrays have different numbers of dimensions, the smaller one is padded with 1s on the LEFT
2. Arrays with size 1 along a dimension act as if they were copied to match the other array
3. If sizes don't match and neither is 1, it's an error

Examples:
```
(3, 4) + (4,)    → (3, 4) + (1, 4) → (3, 4)  ✓  add vector to each row
(3, 4) + (3, 1)  → (3, 4)                      ✓  add column vector to each column
(3, 4) + (3,)    → ERROR                        ✗  3 ≠ 4 and neither is 1
```

````python
# Example: broadcasting in action
M = np.array([[1, 2, 3],
              [4, 5, 6]])

row_vec = np.array([10, 20, 30])        # shape (3,)
col_vec = np.array([[100], [200]])       # shape (2, 1)

print(f"M + row_vec:\n{M + row_vec}")    # adds [10,20,30] to EACH row
print(f"M + col_vec:\n{M + col_vec}")    # adds 100 to row 0, 200 to row 1
````
